```python
import pandas as pd
```

```python
# Read in csv files 
d1 = pd.read_csv('features.csv')
d2 = pd.read_csv('train.csv')

# Combine them
d3 = d1.merge(d2)

# Create new dataframe
d4 = d3[['Store', 'Temperature', 'Weekly_Sales']].copy()
```

```python
# Check if new dataframe is correct
print(d4)
```

```python
d4.plot.line(x='Temperature', y='Weekly_Sales', figsize=(20, 10))
```

```python
print(d4.nlargest(20, 'Weekly_Sales'))
```
